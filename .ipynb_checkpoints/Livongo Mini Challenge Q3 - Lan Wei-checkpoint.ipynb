#### Average avg_bg_value change

```python
pd.DataFrame(df.groupby('date_utc')['avg_bg_value'].mean()).plot(figsize=(10,5))
```

```python
pd.DataFrame(df[(df['gender'] == 'Male')&(df['diabetes_type']=='Type 1')].groupby('date_utc')['avg_bg_value'].mean()).plot(figsize=(10,5))
```

```python
pd.DataFrame(df[(df['gender'] == 'Male')&(df['diabetes_type']=='Type 2')].groupby('date_utc')['avg_bg_value'].mean()).plot(figsize=(10,5))
```

#### People's avg_bg_value is higher than normal during the holiday season. We should send more messages to the members to remind them to keep a healthy lifestyle during the holiday season. Or give them suggestions about what holiday food they can eat what they cannot.

#### Male members average bg_value is higher than female member. So we should send more reminder to male members.


#### Type 1 diabetes patients have higher bg_value compared with type 2 diabetes patients. So they may need more doctor consulting. They also test more bg every day. So they use more test paper and pieces of equipment. So we could give type 1 diabetes members a higher price but better service and provide type 2 diabetes with lower price but more basic service.
